## 1. Correlation Heatmap antar Dimensi Kepercayaan

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example: correlation heatmap between numeric trust dimensions
numeric_cols = [
    'Dimensi Kompetensi',
    'Dimensi Komitmen',
    'Dimensi Transparansi',
    'Dimensi Empati',
    'Keyakinan terhadap Pemerintah'
]

plt.figure(figsize=(6, 4))
sns.heatmap(xlxs[numeric_cols].corr(), annot=True, cmap='Blues', fmt='.2f')
plt.title("Correlation Heatmap antar Dimensi Kepercayaan")
plt.show()
```

## 2. Hubungan Tingkat Pendidikan dengan Persepsi Kompetensi KPK

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=xlxs, x='Tingkat Pendidikan', y='Dimensi Kompetensi')
plt.title("Hubungan Tingkat Pendidikan dengan Persepsi Kompetensi KPK")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

## 3. Persebaran Tingkat Kepercayaan berdasarkan Jenis Kelamin

```python
plt.figure(figsize=(7,4))
sns.countplot(data=xlxs, x='Tingkat Kepercayaan terhadap KPK', hue='Jenis Kelamin')
plt.title("Persebaran Tingkat Kepercayaan berdasarkan Jenis Kelamin")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

## 4. Awareness KPK vs Tingkat Kepercayaan terhadap KPK

```python
import pandas as pd

cross_tab = pd.crosstab(xlxs['Awareness KPK'], xlxs['Tingkat Kepercayaan terhadap KPK'], normalize='index') * 100
cross_tab.plot(kind='bar', stacked=True, figsize=(8,5), colormap='viridis')

plt.title("Hubungan Awareness terhadap Tingkat Kepercayaan KPK")
plt.ylabel("Persentase (%)")
plt.xlabel("Awareness KPK")
plt.legend(title="Tingkat Kepercayaan", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```
