```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('../final/copy-3-response-kpk-survey.xlsx')
```

```python
df.head()
```

```python
# 1.Distribusi Responden Berdasarkan Provinsi (Pertanyaan 4)
print("=" * 70)
print("1. Distribusi Responden Berdasarkan Provinsi")
print("=" * 70)
province_col = "4. Di provinsi manakah Anda berdomisili saat ini?"
province_distribution = df[province_col].value_counts().sort_index()
print(f"\nTotal number of provinces: {df[province_col].nunique()}")
print(f"\nProvince Distribution:\n")
print(province_distribution)
print(f"\nTotal respondents: {len(df)}")
print("\n" + "=" * 70)
```

```python
# visual
plt.figure(figsize=(12, 8))
province_distribution.plot(kind='barh')
plt.title('Distribution of Respondents by Province', fontsize=14, fontweight='bold')
plt.xlabel('Number of Respondents', fontsize=12)
plt.ylabel('Province', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# 2. Total Responden (Pertanyaan 1)
print("=" * 70)
print("2. Total Responden")
print("=" * 70)
name_col = "1. Nama Lengkap"
name_counts = df[name_col].value_counts()
print(f"\nTotal nama (unique): {df[name_col].nunique()}")
print(f"Total responden: {len(df)}")
print(f"\nNama Responden yang sama:\n")
duplicate_names = name_counts[name_counts > 1]
if len(duplicate_names) > 0:
    print(duplicate_names)
    print(f"\nNomor Responden yang sama yang lebih dari 1: {len(duplicate_names)}")
else:
    print("all good no duplicate")
print("\n" + "=" * 70)
```

```python
# 3. Total Responden Berdasarkan Usia (Pertanyaan 3)
print("=" * 70)
print("3. Total Responden Berdasarkan Usia")
print("=" * 70)
age_col = "3. Berapa usia Anda saat ini?"
age_distribution = df[age_col].value_counts().sort_index()
print(f"\nPersebaran Usia:\n")
print(age_distribution)
print(f"\nTotal Responden: {len(df)}")
print(f"\nStatistik Usia:")
print(f"  - Age group: {df[age_col].nunique()}")
print(f"  - Age group paling banyak: {age_distribution.index[0]} ({age_distribution.iloc[0]} responden)")
print(f"  - Age group paling sedikit: {age_distribution.index[-1]} ({age_distribution.iloc[-1]} responden)")
print("\n" + "=" * 70)
```

```python
# Visual
plt.figure(figsize=(10, 7))
age_distribution.plot(kind='bar', color='steelblue')
plt.title('Total Responden Berdasarkan Usia', fontsize=14, fontweight='bold')
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Number of Respondents', fontsize=12)
plt.xticks(rotation=0, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
province_col = "4. Di provinsi manakah Anda berdomisili saat ini?"
name_col = "1. Nama Lengkap"

# Aggregate names into a list with proper column naming
province_summary = (
    df.groupby(province_col)[name_col]
      .agg(Count='count', Names=list)
      .reset_index()
      .rename(columns={province_col: 'Province'})
      .sort_values(by='Count', ascending=False)
)

print("=" * 70)
print("Summary of Achieved Respondents by Province")
print("=" * 70)
print(f"\nTotal Respondents: {province_summary['Count'].sum()}")
print("\n")

# Pretty print per province
for _, row in province_summary.iterrows():
    print(f"ğŸ“ Province: {row['Province']}  â€”  {row['Count']} respondents")
    print("   Names:")
    
    for name in row['Names']:
        print(f"     â€¢ {name}")
    
    print("-" * 70)
```
